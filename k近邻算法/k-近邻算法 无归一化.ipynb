# 第二章 k-近邻算法 无归一化

@ 分类 和 回归 属于 监督学习

@ 无监督学习 可以 减少 数据特征 的 维度，可以使用 二维 或 三维图形 直观地 展示 数据信息。

@ 简单说，k-近邻算法 采用 测量不同特征值之的距离 方法 进行分类

@ 计算样本集中 所有样本 与 未知样本 的 距离， 取 离最近的 前k个，根据其类型来决定 位置样本 的 类型

```python
from collections import Counter

def k_Nearest_Neighbor(group,labels,new,k):
    all_error = []
    for i,j in zip(group,labels):
        sum_sq = sum([(i[k]-new[k])**2 for k in range(len(i))])
        error = sum_sq**0.5
        all_error.append([error,j])
    #print("all_error\n",all_error)
    all_error.sort()
    #print("sorted all_error\n",all_error)
    #print("all_error[:k]\n",all_error[:k])
    
    first_k_labels = [i[1] for i in all_error[:k]]
    print("first_k_labels\n",first_k_labels)
        
    label_counter = Counter(first_k_labels)
    
    #label_counter = {"a":2,"b":4,"c":1,"d":4}
    print("label_counter\n",label_counter)
    max_value = max(label_counter.values())
    print("max_value\n",max_value)
    
    ##### IMPORTANT! #####
    # max_label = max(label_counter, key=lambda key: label_counter[key])
    # print("max_label\n",max_label)
    
    max_labels = []
    for key,value in label_counter.items():
        if value == max_value:
            max_labels.append(key)
 
    print("max_labels\n",max_labels)
    return max_labels
```

```python
# Simple test
group = [[1,1.1],[1,1],[0,0],[0,0.1]]
labels = ["A","A","B","B"]
k_Nearest_Neighbor(group,labels,[1,2],2)
```

```python
# Input
group = []
labels = []
with open("datingTestSet.txt") as f:
    for i in f:
        instance = i.split()
        for i in range(3):
            instance[i] = float(instance[i])
        group.append(instance[:3])
        labels.append(instance[3])
print(group[:5])
print(labels[:5])
```

```python
# Real test
rst = []
for i in group[:100]:
    tmp = k_Nearest_Neighbor(group,labels,i,30)
    rst.append(tmp)
```

```python
# Standard kNN
from numpy import *
import operator

def classify0(inX, dataSet, labels, k):
    dataSetSize = array(dataSet).shape[0]
    diffMat = tile(inX, (dataSetSize,1)) - dataSet
    sqDiffMat = diffMat**2
    sqDistances = sqDiffMat.sum(axis=1)
    distances = sqDistances**0.5
    sortedDistIndicies = distances.argsort()     
    classCount={}          
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]]
        classCount[voteIlabel] = classCount.get(voteIlabel,0) + 1
    sortedClassCount = sorted(classCount.items(), key=operator.itemgetter(1), reverse=True)
    return sortedClassCount[0][0]
```

```python
# test
rst1 = []
for i in group[:100]:
    tmp = classify0(i,group,labels,30)
    rst1.append(tmp)
```

```python
# Compare
difference_count = 0
for i in range(100):
    print(i,rst[i],rst1[i])
    if rst[i][0]!= rst1[i] or len(rst[i])!=1:
        difference_count+=1
difference_count
```
